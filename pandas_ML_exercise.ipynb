good for classification but we want our feature to preserve meaning

hard code values in a dict

```python
cut_class_dict = {"Fair": 1, "Good": 2, "Very Good": 3, "Premium": 4, "Ideal": 5}
```

```python
df['clarity'].unique()
```

```python
clarity_dict = {"I3": 1, "I2": 2, "I1": 3, "SI2": 4, "SI1": 5, "VS2": 6, "VS1": 7, "VVS2": 8, "VVS1": 9, "IF": 10, "FL": 11}
color_dict = {"J": 1,"I": 2,"H": 3,"G": 4,"F": 5,"E": 6,"D": 7}
```

```python
df['cut'] = df['cut'].map(cut_class_dict)
df['clarity'] = df['clarity'].map(clarity_dict)
df['color'] = df['color'].map(color_dict)
df.head()
```

```python
import sklearn
from sklearn import svm, preprocessing

df = sklearn.utils.shuffle(df)

X = df.drop("price", axis=1).values  # feature set
X = preprocessing.scale(X)
y = df['price'].values # label

test_size = 200

X_train = X[:-test_size]
y_train = y[:-test_size]

X_test = X[-test_size:]
y_test = y[-test_size:]

clf = svm.SVR(kernel="linear")
clf.fit(X_train, y_train)
```

to prevent bias in model building, the data must be shuffed if it exists in any sorted pattern

the last value passed to the model will effect output, so shuffle

```python
X
```

```python
clf.score(X_test, y_test)
```

```python
for X, y in zip(X_test, y_test):
    print(f"Model: {clf.predict([X])[0]}, Actual: {y}")
```

```python
clf = svm.SVR(kernel="rbf")
clf.fit(X_train, y_train)
```

```python
print(clf.score(X_test, y_test))

for X, y in zip(X_test, y_test):
    print(f"Model: {clf.predict([X])[0]}, Actual: {y}")
```
