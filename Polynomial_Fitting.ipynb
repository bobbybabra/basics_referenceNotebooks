## Plynomial Fitting with NumPy polyfit

Simple Line: Polynomial of Degree 1
    
$$y = m * x + b$$
    
Quadriatic equation: Degree 2
    
$$y = ax^{2} + bx + c$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-10, 10, 10)
y = 2 * x + 5

plt.plot(x, y, 'o')
```

```python
poly_coeff = np.polyfit(x, y, 2) # polyfiting degree 2 to get 3 coeffecients
poly_coeff
```

```python
xnew = np.linspace(-20,20, 100)
ynew = np.poly1d(poly_coeff)  # y is assigned to 1st degree poly with calculated coeff

plt.plot(xnew, ynew(xnew), x, y, 'o')
```

###### Finding a value of the fxn at any point

```python
ynew(0.5)
```

```python
ynew = y + 5 * np.random.rand(10)
plt.plot(x, ynew, 'r-*')
```

```python
pol_coef = np.polyfit(x, ynew, 2)
yfit = np.poly1d(pol_coef)
plt.plot(xnew, yfit(xnew), x, ynew, 'go')
```
